```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head(2)
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head(2)
```

```python
movies = movies.merge(credits, on = 'title')
movies.head(2)
```

```python
movies['original_language'].value_counts()
```

```python
movies = movies[['movie_id_x', 'title', 'overview', 'genres', 'keywords', 'cast_y', 'crew_y']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```
